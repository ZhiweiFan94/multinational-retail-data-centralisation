# Users dataset 

```python
#clean the users data
import pandas as pd
import numpy as np

df = pd.read_csv('users.csv')
df.head(5)
```

```python
df.isna().sum()
```

```python
df.count()
```

```python
df.dropna(inplace=True)
df.count()
```

```python
df.isna().sum()
```

```python
#ok now the easiest thing to check is the country code, as it should be the known info.
df['country_code'].unique()
```

```python
# let us see what is GGB
df[df['country_code'] == 'GGB']
```

```python
# as it is also united kingdom, so it must be typo, revise it to GB
df['country_code'] = df['country_code'].apply(lambda x: 'GB' if x=='GGB' else x)
```

```python
df['country_code'].unique()
```

```python
# alright, next is to drop all invalid country code from dataset
df = df[df['country_code'].isin(['DE', 'GB', 'US'])]
```

```python
#let us examine if the useless country code infos are removed
df.count()
```

```python
#lets check the DoB, if they are write in the same style
#use str.match method 
#define the patter xxxx-xx-xx Year-month-day
pattern = r'^\d{4}-\d{2}-\d{2}$'
# Check if each entry in the 'date_of_birth' column matches the pattern
matches_format = df['date_of_birth'].astype(str).str.match(pattern)
print(matches_format)
```

```python
sum(matches_format)
#it means 15320-15257 is the format not in the right order
```

```python
# let's check if the DoB is valid or just useless
print(df[df['date_of_birth'].astype(str).str.match(pattern) == False]['date_of_birth'])
```

```python
#alright, then the DoB are valid let use pd.to_datetime() method
df['date_of_birth_revised'] = pd.to_datetime(df['date_of_birth'])
df['date_of_birth_revised'].loc[13045]
```

```python
#double check to see if all match the right format, alright, we are good!
print(df[df['date_of_birth_revised'].astype(str).str.match(pattern) == False])
```

```python
df_file.isna().sum()
```
